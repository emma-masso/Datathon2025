```python
# Codi per adaptar l'archiu CSV de test
# Per cada fila del CSV original, crear tantes files com el valor de life_cycle_length, una per cada setmana
import pandas as pd

df = pd.read_csv("./../data/test.csv",sep=";")

# Aqui eliminen les columnes innecessaries
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
df.head(20)
```

```python
# Aqui expandim les files segons life_cycle_length
df_expanded = df.loc[df.index.repeat(df['life_cycle_length'])].reset_index(drop=True)

# Crear columna num_week_iso numerant cada copia desde 1 fins a n
df_expanded['num_week_iso'] = df_expanded.groupby('ID').cumcount() + 1
df_expanded["year"]=2025
df_expanded["weekly_demand"]=0 # Afegim columna per modificar-la despres amb les prediccions

        
df_expanded.head(20)
```

```python
df_expanded.to_csv("./../data/test-converted.csv", index=False)
```
