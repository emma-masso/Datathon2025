```python
# Amb això creem un nou CSV per fer la submissió 
import pandas as pd

dfp = pd.read_csv("./../data/test-parsed-with-predictions-rf.csv",sep=",") # D'aqui traiem les prediccions
#dfo = pd.read_csv("./data/train-parsed.csv",sep=",")
dfID = pd.read_csv("./../data/test-converted.csv",sep=",") # D'aqui traiem els ID

#dfo["ID"] = dfID["ID"]
dfp["ID"] = dfID["ID"]
```

```python
# Calculem la suma per ID, per calcular la demanda total per producte
dfp_sum = dfp.groupby('ID', as_index=False)['weekly_demand'].sum()
```

```python
dfp_sum.head(100)
```

```python
file_sortida = "./../data/test-results3.csv"

dfp_sum["ID"] = dfp_sum["ID"].astype(int)

dfp_sum["Production"] = dfp_sum["weekly_demand"]
dfp_sum["Production"] = dfp_sum["Production"].round().astype(int)
dfp_sum.drop(columns=["weekly_demand"], inplace=True)
dfp_sum.to_csv(file_sortida, index=False)
dfp_sum.head(100)
```
